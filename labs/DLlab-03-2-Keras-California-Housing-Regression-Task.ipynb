Kvetkez lpsknt importljuk a szksges modulokat s lltsunk be egy random seed-et:

````python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.models import load_model
import copy
import pandas as pd
from tensorflow.keras.optimizers import SGD
from sklearn.preprocessing import StandardScaler


np.random.seed(123)
````

Ezutn tltsk le s tltsk be az adatokat s vizsgljuk meg hogy milyen oszlopokat (bemen vltozk s cl/becslt vltoz) tartalmaz.

````python
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)
````

````python
california_housing.frame
````

Vlasszuk szt a model be- s kimenett jelent vltozkat,  train-validation-test adathalmazokat s standardizljuk a bemeneteket

````python
dataset = california_housing.frame.values
num_samples = dataset.shape[0]
test_split = 0.1
valid_split = 0.1

# Be s kimenetek sztvlasztsa
#X teljes adathalmaz bemen vltozi
#Y teljes adathalmaz cl vltozja (aminek becslsre a hlt tantani akarjuk, elvrt kimenetek)
X, Y = dataset[:, 0:8], dataset[:, 8]


# Adatok sztvlasztsa train-validation-test partcikra
X_test = X[int(num_samples * (1-test_split)):]
Y_test = Y[int(num_samples * (1-test_split)):]
X_valid = X[int(num_samples * (1 - valid_split - test_split)):int(num_samples * (1-test_split))]
Y_valid = Y[int(num_samples * (1 - valid_split - test_split)):int(num_samples * (1-test_split))]
X = X[0:int(num_samples * (1 - valid_split - test_split))] #tant bemenetek (lustn neveztk el a vltozt, X_train lenne a beszdesebb)
Y = Y[0:int(num_samples * (1 - valid_split - test_split))] # tant kimenetek (Y_train...)

# Standardizls
scaler = StandardScaler().fit(X)
X = scaler.transform(X)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
````

````python
X.shape, Y.shape
````

````python
X_test.shape, Y_test.shape
````

````python
X_valid.shape, Y_valid.shape
````

Ezt kveten adjuk meg a callback fggvnyeket, amelyek biztostjk, hogy:
- A tants lljon le, ha a 40 epochon t nem javul a validcis hiba
- A legjobb modelt elmentjk a `weights.hdf5` fjlba.

````python
# callback-ek definilsa
early_stopping = EarlyStopping(patience=40, verbose=1)

checkpointer = ModelCheckpoint(filepath='weights.hdf5', save_best_only=True, verbose=1)
````

Hozzuk ltre a hlt s tantsuk be!
- A hl 2 egyszer neuron rtegbl lljon
    1. rteg
        - bemeneteinek szma a bemen vltozk szma
        - 200 neuront tartalmazzon
        - Szigmoid aktivcis fggvny
    2. rteg
        - ez adja a kimenetet, teht a clvltozk szmval egyezen meg a neuronok szma (units)
        - aktivcis fggvnynek gondold t mit rdemes ehhez a feladathoz hasznlni(`linear`, `sigmoid`, `tanh`, ...)
- Optimaliccihoz SGD algoritmust hasznlj, $10^{-3}$ tanulsi rtval,  $10^{-6}$ L2 regularizci egytthatval (weight decay), valamint a kvetkez momentum belltssal: `momentum=0.9, nesterov=True`
- Hibafggvny: ngyzetes hiba
- Batch mret: 16

> Ha sikerlt a tants kdjt sszerakni elindtani, akkor trj ide vissza s mdostsd, a kt rteg kz adj hozz egy Dropout(0.5) rteget!

````python
model = Sequential()
model.add(Dense(200, input_shape=(X.shape[1], ), activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='relu'))

sgd = SGD(
    learning_rate=10e-3,
    weight_decay=10e-6,
    momentum=0.9,
    nesterov=True
)

model.compile(
    loss='mse',
    optimizer=sgd
)
````

````python
history= model.fit(
    X, Y,
    validation_data=(X_valid, Y_valid),
    batch_size=16,
    callbacks=[
        early_stopping,
        checkpointer
    ]
)
````

A tants vgeztvel visszatltjk a legjobb modellt, s kirtkeljk ennek teljestmnyt a teszt adatokon. Mit jelenthet ez az rtk?

````python
model = load_model('weights.hdf5') # a 'weights.hdf5' fjlba mentett legjobb modell visszatltse

from sklearn.metrics import mean_squared_error
import math

preds = model.predict(X_test)# predikcik szmtsa a teszt adatokon a betlttt modellel
test_err = mean_squared_error(Y_test, preds)# szmts ngyzetes hibt a teszt predikcik s az elvrt kimenetek kztt
print("\nTeszt hiba: %f" % (test_err))
print(f"Ez az jelenti, hogy sqrt({test_err:0.3f}))={math.sqrt(test_err):0.3f}-at hibz tlagosan a modell a teszt adatokon.")
print(f"Ez annyit jelent, hogy tlagosan {math.sqrt(test_err)*100000:0.0f}$-t tved a modellnk.")
````
