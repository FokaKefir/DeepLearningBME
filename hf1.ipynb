# 2. feladat

**Átlag (mean)**: Az átlag az adatok összegzése és az adatok számának osztásával számolható ki. Az átlag azt mutatja, hogy az adatok közötti középpont hol helyezkedik el. Általánosan egy adathalmaz átlagát számolva az adatokat egyenletesen osztjuk el, ami azt eredményezi, hogy az átlag értéke közelíti az adatok középpontját. Például egy színkép esetén az átlagot minden pixel értékére számolhatjuk az összes pixelen, és ez az átlag a kép "átlagos" színét jelenti.

**Szórás (standard deviation - std)**: A szórás azt mutatja, hogy az adatok mennyire terjednek el az átlagtól. Minél nagyobb a szórás, annál nagyobb az adatok szóródása az átlagtól. A szórás számításához az adatokat az átlagtól való eltéréseik négyzetének átlagát veszik. Szórás nélkül az adathalmazok közötti különbségeket nem lehetne megfigyelni vagy összehasonlítani. A képnormalizálás során a szórás segít abban, hogy a pixelek értékei "stabilizálódjanak" egy adott skálán, amely könnyen kezelhető a további feldolgozás során.

A képnormalizálás során az átlag és a szórás használata segít abban, hogy a képeket azonos skálán mozgasd (normalizáld) oly módon, hogy az átlag 0 legyen, és a szórás pedig 1. Ezáltal az adatok könnyen kezelhetők és értelmezhetők, és a gépi tanulási vagy mélytanulási modellek jobban teljesíthetnek. Az ilyen normalizált adathalmazokat könnyebb összehasonlítani és tanítani a gépi tanulási algoritmusoknak.

Az `eps`-ra azért van szükség mert a szórás értéke a harmadik kép zöld csatornáján pont 0, ezáltal az osztás eredménye `nan` lenne. Az `eps` egy borzasztó kicsi érték, ami épp, hogy csak annyira tolja el a szórást, hogy az osztás eredményén ne lehessen megfigyelni a különbséget és mégis elvégezhető legyen a művelet.

```python
# epsilon is a small number close to 0
eps = 1e-9

# Go through every image
for index, img in enumerate(imgs):
  print(f'img{index + 1}')

  # Calculate mean and std for every color channel
  means = np.mean(img, axis=(0, 1))
  stds = np.std(img, axis=(0, 1))

  print("means:", means)
  print("stds:", stds)

  # Add epsilon to the every std, so we can avoid dividing with 0
  stds += eps

  # Normalize the spectrogram data
  norm_img = (img - means) / stds

  # Print normalized image
  print(norm_img)

  # Check if the values are correct
  print(f"Mean of normalized image: {np.mean(norm_img)}") # Mean should be close to 0
  print(f"Standard deviation of image: {np.std(norm_img)}") # Std should be close to 1
  print()
```

```python
# Import dependencies
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import numpy as np
```

```python
# Make a function for plotting the informations about audio signal
def plot_audio_signal_data(audio_file):
  # Read audio data from file
  sample_rate, samples = wavfile.read(audio_file)

  # Calculate informations for spectogram
  freq, times, spectrogram = signal.spectrogram(samples, sample_rate)

  # Plot the audio waveform
  plt.figure(figsize=(12, 10))
  plt.subplot(2, 1, 1)
  plt.ylabel('Amplitude [dB]')
  plt.xlabel('Time [sec]')
  plt.title('Audio Waveform')
  plt.plot(np.arange(len(samples)) / sample_rate, samples)

  # Plot the spectrogram
  plt.subplot(2, 1, 2)
  plt.ylabel('Frequency [Hz]')
  plt.xlabel('Time [sec]')
  plt.title('Spectrogram')
  plt.pcolormesh(times, freq, 10 * np.log10(spectrogram))

  plt.suptitle(f'Audio Analysis for {audio_file}', fontsize=16)
  plt.show()
```

```python
plot_audio_signal_data('audio1.wav')
```

```python
plot_audio_signal_data('audio2.wav')
```

```python
# Make function that normalize audio file's spectogram
def normalized_audio_signal_data(audio_file):
  # Load the audio file
  sample_rate, samples = wavfile.read(audio_file)

  # Compute the spectrogram
  freq, times, spectrogram = signal.spectrogram(samples, sample_rate)

  # Normalize the spectrogram data
  mean = np.mean(spectrogram)
  std_dev = np.std(spectrogram)
  normalized_spectrogram = (spectrogram - mean) / std_dev

  # Check if normalization is correct
  print(f"Mean of normalized spectrogram: {np.mean(normalized_spectrogram)}")
  print(f"Standard deviation of normalized spectrogram: {np.std(normalized_spectrogram)}")
```

```python
normalized_audio_signal_data('audio1.wav')
```

```python
normalized_audio_signal_data('audio2.wav')
```

```python
# Import dependecies
import requests
from bs4 import BeautifulSoup
import re
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# scrape the data from url
url = "http://smartlab.tmit.bme.hu/oktatas-deep-learning"
response = requests.get(url)

if response.status_code == 200:
  # Parse the HTML content
  soup = BeautifulSoup(response.text, "html.parser")

  # Extract the text from the HTML content
  text = soup.get_text()

  # Display the text
  #print("Page content:")
  #print(text)

  # Remove non-alphabetic characters and convert to lowercase
  text = re.sub(r'[^a-zA-Z]', '', text).lower()

  # Count the frequency of letters
  letter_counts = Counter(text.lower())

  # Sort the letters alphabetically
  sorted_letter_counts = sorted(letter_counts.items())

  # Extract the sorted letters and frequencies
  letters, frequencies = zip(*sorted_letter_counts)

  # Display the histogram
  plt.bar(letters, frequencies)
  plt.xlabel("Letters")
  plt.ylabel("Frequency")
  plt.title("Frequency of Letters")
  plt.show()

else:
  print(f"Error while downloading the web page. Status code: {response.status_code}")
```
